## Part 1

```python
irisFileURL = 'iris.data'
irisData = pd.read_csv(irisFileURL, names=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width', 'Name'], header=None)
```

### 1.1 Summary Statistics

```python
# The first 5 elements of the Iris dataset
irisData.head(5)
# There four features: Sepal Length, Sepal Width, Petal Length, Petal Width
# One class: Name
# features are all numeric 
# class is nominal
```

```python
for attribute in irisData:
    if attribute is not 'Name':
        print("The minimum value of {}: {}".format(attribute, irisData[attribute].min()))
        print("The maximum value of {}: {}".format(attribute, irisData[attribute].max()))
        print("The mean of {}: {}".format(attribute, irisData[attribute].mean()))
        print("The range of {}: {}".format(attribute, irisData[attribute].ptp()))
        print("The standard deviation of {}: {}".format(attribute, irisData[attribute].std()))
        print("The variance of {}: {}".format(attribute, irisData[attribute].var()))
        print("The count of Sepal Length: {}: {}".format(attribute, irisData[attribute].count()))
        print("The 25:50:75% percentiles of {}:\n{}".format(attribute, irisData[attribute].quantile([0.25, 0.5, 0.75])))
        print("\n")
    else:
        print(irisData[attribute].describe())
```

### 1.2 Data Visualization

```python
irisData.hist(figsize=(15, 8));
```

```python
irisData.plot(kind="box",figsize=(15, 8));
```

```python
pbhdFileURL = 'pendigits.tra'
pbhdData = pd.read_csv(pbhdFileURL, header=None)
pbhdData.head(5)
# There are 16 features and one class
# all features are numeric from 0 to 100 
# class is numberic code from 0 to 9
```

```python
# Summary Statistic of PBHD Data for each feature and class
for column in pbhdData:
    print("The minimum value of column {}: {}".format(column, pbhdData[column].min()))
    print("The maximum value of column {}: {}".format(column, pbhdData[column].max()))
    print("The mean of column {}: {}".format(column, pbhdData[column].mean()))
    print("The range of Petal column {}: {}".format(column, pbhdData[column].ptp()))
    print("The standard deviation of column {}: {}".format(column, pbhdData[column].std()))
    print("The variance of column {}: {}".format(column, pbhdData[column].var()))
    print("The count of column {}: {}".format(column, pbhdData[column].count()))
    print("The 25:50:75% percentiles of column {}: \n{}".format(column, pbhdData[column].quantile([0.25, 0.5, 0.75])))
    print("\n")
```

```python
petalLength = irisData['Petal Length']
# To better understanding the histogram of Iris Data
# Consider the following
irisData.hist();
'The Petal Length and Petal Width both have a short gap where the Sepal Length and Sepal Width are continuous.'\
'The later twos have a normal shape where the first two have a irregular shape.'\
'The range of Petal Length is relatively larger (5.9) where the range of Sepal Length is 3.6.'
```

```python
print('If there were a particular value in Petal Length to separate the Petal Length graph, it would be 2.5.\n'\
'Because according to the graph, there is no data to be plotted there.\n'\
'Once it gets separated, the two segments make a little more sense than the original one.\n'\
'For instance, compare the following mean and median.\n')

print("Originally, The mean of Petal Length: {}, the median of Petal Length: {}".format(petalLength.mean(), petalLength.median()))
seg1 = petalLength[petalLength <= 2.5]
seg2 = petalLength[petalLength > 2.5]
print("Segment 1: The mean of Petal Length: {}, the median of Petal Length: {}".format(seg1.mean(), seg1.median()))
print("Segment 2: The mean of Petal Length: {}, the median of Petal Length: {}".format(seg2.mean(), seg2.median()))

'After separated, the mean and median are much closer than the orginal one.'
```
